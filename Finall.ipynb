```python
# TODO: ุงูุฏุงุฒู ุชุตุงูุฑ ู ุงูุฏุงุฒู ูุฑ ุจฺ (ุชุนุฏุงุฏ ุชุตุงูุฑ ุฏุฑ ูุฑ ุฏุณุชู ุขููุฒุด) ุฑุง ูุดุฎุต ฺฉูุฏ
import tensorflow as tf
from tensorflow.keras import layers


BASE_DIR = '/content/drive/MyDrive/rps/rps_dataset'


# TODO: ุงูุฏุงุฒู ุชุตุงูุฑ ู ุงูุฏุงุฒู ูุฑ ุจฺ
IMG_SIZE   = (224, 224)   # ูโุชูู (200,300) ูู ุจุฐุงุฑ ูู 224 ุงุณุชุงูุฏุงุฑุฏุชุฑู
BATCH_SIZE = 32

VAL_SPLIT = 0.20
SEED      = 42

# ุฏุงุฏูโูุง ุขููุฒุด
train_ds = tf.keras.utils.image_dataset_from_directory(
    BASE_DIR,
    validation_split=VAL_SPLIT,
    subset="training",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode="categorical"   # one-hot
)

# ุฏุงุฏูโูุง ุงุนุชุจุงุฑุณูุฌ
val_ds = tf.keras.utils.image_dataset_from_directory(
    BASE_DIR,
    validation_split=VAL_SPLIT,
    subset="validation",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode="categorical"
)
```

```python
# TODO: ููุงุด ุฏุงุฏูโูุง Augmented
import math

# TODO: ููุงุด ุฏุงุฏูโูุง Augmented
def visualize_augmented_images(dataset, augmentation_layer, num_samples=9):
    """
    ููุงุด ุชุนุฏุงุฏ ุชุตูุฑ ุงูุฒุงุดโ ุงูุชู (Augmented) ุงุฒ ฺฉ ุฏุชุงุณุช ุขููุฒุด.
    """
    # ฺฉ batch ุงุฒ ุฏุชุงุณุช ุจฺฏุฑ
    images, labels = next(iter(dataset))
    batch_size = images.shape[0]

    # ุชุนุฏุงุฏ ูููููโูุง ฺฉู ููุงุด ูโุฏูู ุฑุง ุจู ุงูุฏุงุฒูโ batch ูุญุฏูุฏ ฺฉู
    n = min(num_samples, int(batch_size))

    # ุงุนูุงู ุงูุฒุงุด ุฏุงุฏู (training=True ุชุง ุชุจุฏูุงุช ุชุตุงุฏู ูุนุงู ุดููุฏ)
    aug = augmentation_layer(images, training=True)

    # ุชุจุฏู ุงูู ุจุฑุง ููุงุด (ุจู uint8 ุฏุฑ ุจุงุฒูโ 0..255)
    def to_uint8(x):
        x = tf.convert_to_tensor(x)
        # ุงฺฏุฑ ุฏุงุฏูโูุง ุฏุฑ ุจุงุฒูโ 0..1 ุจุงุดูุฏุ ุถุฑุจ ุฏุฑ 255
        xmax = tf.reduce_max(x)
        if tf.math.less_equal(xmax, 1.0):
            x = x * 255.0
        x = tf.clip_by_value(x, 0.0, 255.0)
        return tf.cast(x, tf.uint8)

    aug = to_uint8(aug)

    # ุงุจุนุงุฏ ฺฏุฑุฏ (ูุฒุฏฺฉ ูุฑุจุน)
    cols = int(math.ceil(math.sqrt(n)))
    rows = int(math.ceil(n / cols))

    plt.figure(figsize=(3*cols, 3*rows))
    for i in range(n):
        ax = plt.subplot(rows, cols, i + 1)
        plt.imshow(aug[i].numpy())
        plt.axis('off')
    plt.suptitle("Augmented Samples", y=0.98)
    plt.tight_layout()
    plt.show()

visualize_augmented_images(train_ds, data_augmentation, num_samples=9)
```

```python
# ุชุนุฑู ุชุงุจุน ูพุดโูพุฑุฏุงุฒุด ุดุงูู ูุฑูุงูโุณุงุฒ ููุงุฏุฑ ูพฺฉุณู
def preprocess(image, label):
    # ุชุจุฏู ููุน ุฏุงุฏู ุจู float32
    image = tf.cast(image, tf.float32)
    # ูุฑูุงูโุณุงุฒ ุจู ุจุงุฒู [0,1]
    image = image / 255.0
    return image, label

# ุงุนูุงู ุฑู ุฏุชุงุณุช ุขููุฒุด ู ุงุนุชุจุงุฑุณูุฌ
train_ds = train_ds.map(preprocess, num_parallel_calls=tf.data.AUTOTUNE)
val_ds   = val_ds.map(preprocess, num_parallel_calls=tf.data.AUTOTUNE)
```

```python
# TODO: ุณุงุฎุช ูุฏู ุดุจฺฉู ุนุตุจ ฺฉุงููููุดู (CNN) ุจุฑุง ุทุจููโุจูุฏ ุชุตุงูุฑ

import tensorflow as tf
from tensorflow.keras import layers, models, optimizers

NUM_CLASSES = len(class_names)
INPUT_SHAPE = IMG_SIZE + (3,)

# --- Augmentation ูพุดููุงุฏ ููโุชุฑ ---
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.2),
    layers.RandomZoom(0.15),
    layers.RandomContrast(0.15),
    layers.RandomTranslation(0.1, 0.1),
], name="data_augmentation")

# --- MobileNetV2 ูพุงู (ImageNet) ---
base = tf.keras.applications.MobileNetV2(
    input_shape=INPUT_SHAPE, include_top=False, weights="imagenet"
)
base.trainable = False  # ูุฑุญูู ุงูู: ูุฑุฒ

# --- ูุฏู ููุง ---
inputs = layers.Input(shape=INPUT_SHAPE)
x = data_augmentation(inputs)

# ุงฺฏุฑ ุชุตุงูุฑุช ุงุฒ ูุจู ุฏุฑ ุจุงุฒู [0,1] ูุณุชูุฏุ ุงู ูุงู ุขู ุฑุง ุจู [-1,1] ูโุจุฑุฏ
# (ุจุง ุงู ฺฉุงุฑ ูุงุฒ ุจู mobilenet_v2.preprocess_input ูุณุช ู ุชุฏุงุฎู ูู ุจุง pipeline ุชู ูพุด ููโุขุฏ)
x = layers.Lambda(lambda t: (tf.cast(t, tf.float32) * 2.0) - 1.0, name="to_minus1_1")(x)

x = base(x, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.4)(x)
outputs = layers.Dense(NUM_CLASSES, activation="softmax")(x)

model = models.Model(inputs, outputs, name="rps_mobilenetv2_tl")

```

```python
import matplotlib.pyplot as plt

def plot_training_history(history):
    """
    ุฑุณู ฺฏุฑุงูโูุง ุฏูุช ู ุฎุทุง ุจุฑุง ุฏุงุฏูโูุง ุขููุฒุด ู ุงุนุชุจุงุฑุณูุฌ.
    """
    acc      = history.history['accuracy']
    val_acc  = history.history['val_accuracy']
    loss     = history.history['loss']
    val_loss = history.history['val_loss']
    epochs   = range(1, len(acc) + 1)

    # --- ูููุฏุงุฑ ุฏูุช ---
    plt.figure(figsize=(12,5))

    plt.subplot(1,2,1)
    plt.plot(epochs, acc, 'b-', label='Training Accuracy')
    plt.plot(epochs, val_acc, 'r-', label='Validation Accuracy')
    plt.title('Accuracy over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)

    # --- ูููุฏุงุฑ ุฎุทุง ---
    plt.subplot(1,2,2)
    plt.plot(epochs, loss, 'b-', label='Training Loss')
    plt.plot(epochs, val_loss, 'r-', label='Validation Loss')
    plt.title('Loss over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

history = model.fit(train_ds, validation_data=val_ds, epochs=30, callbacks=callbacks)
plot_training_history(history)
```

```python
import os, random

def predict_random_samples(model, base_dir, class_names, k=2, img_size=(224,224), seed=None):
    """
    Show k random samples per class in a grid (rows=classes, cols=k), with model predictions.
    """
    if seed is not None:
        random.seed(seed)

    rows, cols = len(class_names), k
    plt.figure(figsize=(3*cols, 3*rows))

    for r, cls in enumerate(class_names):
        folder = os.path.join(base_dir, cls)
        files = [f for f in os.listdir(folder)
                 if f.lower().endswith((".jpg", ".jpeg", ".png", ".bmp", ".gif"))]

        if not files:
            ax = plt.subplot(rows, cols, r*cols + 1)
            ax.text(0.5, 0.5, f"No images in {cls}", ha="center", va="center")
            ax.axis("off")
            continue

        picks = random.sample(files, min(k, len(files)))

        for c, fname in enumerate(picks):
            path = os.path.join(folder, fname)

            # load & preprocess
            img = tf.keras.utils.load_img(path, target_size=img_size)
            arr = tf.keras.utils.img_to_array(img) / 255.0
            batch = arr[None, ...]  # (1,H,W,C)

            # predict
            probs = model.predict(batch, verbose=0)[0]
            idx = int(np.argmax(probs))
            pred = class_names[idx]
            conf = float(probs[idx]) * 100.0

            # plot
            ax = plt.subplot(rows, cols, r*cols + c + 1)
            ax.imshow(img); ax.axis("off")
            ax.set_title(f"True: {cls}\nPred: {pred} ({conf:.1f}%)", fontsize=10)

    plt.suptitle(f"{k} random samples per class", y=0.995, fontsize=12)
    plt.tight_layout()
    plt.show()



predict_random_samples(model, "/content/drive/MyDrive/rps/rps_dataset",
                             class_names, k=3, img_size=IMG_SIZE, seed=42)   # 9 images
```
