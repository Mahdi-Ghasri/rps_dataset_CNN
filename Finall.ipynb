```python
# TODO: تابع نمایش تصاویر رندوم را کامل کنید.
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def visualize_two_sample_images(directory, title="Sample Images"):
    """
    دو تصویر تصادفی از مسیر مشخص‌ شده بارگذاری و نمایش می‌دهد.

    پارامترها:
    directory (str): مسیر پوشه‌ای که تصاویر در آن قرار دارند.
    title (str): عنوانی که بالای هر تصویر نمایش داده می‌شود. (پیش‌فرض: "Sample Images")
    """
    # همه فایل‌ها
    images = os.listdir(directory)
    # انتخاب 2 تا رندوم
    samples = random.sample(images, 2)

    plt.figure(figsize=(6, 3))
    for i, img_name in enumerate(samples, 1):
        img_path = os.path.join(directory, img_name)
        img = mpimg.imread(img_path)

        plt.subplot(1, 2, i)
        plt.imshow(img)
        plt.axis("off")
        plt.title(title)

    plt.tight_layout()
    plt.show()


visualize_two_sample_images(PAPER_DIR, title="Paper")
visualize_two_sample_images(ROCK_DIR, title="Rock")
visualize_two_sample_images(SCISSORS_DIR, title="Scissors")
```

```python
# TODO: اندازه تصاویر و اندازه هر بچ (تعداد تصاویر در هر دسته آموزش) را مشخص کنید
import tensorflow as tf
from tensorflow.keras import layers


BASE_DIR = '/content/drive/MyDrive/rps/rps_dataset'


# TODO: اندازه تصاویر و اندازه هر بچ
IMG_SIZE   = (224, 224)   # می‌تونی (200,300) هم بذاری ولی 224 استانداردتره
BATCH_SIZE = 32

VAL_SPLIT = 0.20
SEED      = 42

# داده‌های آموزش
train_ds = tf.keras.utils.image_dataset_from_directory(
    BASE_DIR,
    validation_split=VAL_SPLIT,
    subset="training",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode="categorical"   # one-hot
)

# داده‌های اعتبارسنجی
val_ds = tf.keras.utils.image_dataset_from_directory(
    BASE_DIR,
    validation_split=VAL_SPLIT,
    subset="validation",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode="categorical"
)
```

```python
# TODO: افزایش داده

from tensorflow.keras import layers
import tensorflow as tf

data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.2),
], name="data_augmentation")

data_augmentation


images, labels = next(iter(train_ds))


augmented = data_augmentation(images, training=True)

# نمایش 6 تصویرِ تقویت‌شده
n = 6
plt.figure(figsize=(12, 4))
for i in range(n):
    ax = plt.subplot(2, 3, i + 1)
    # تبدیل به محدوده و نوع مناسب برای نمایش
    img = tf.clip_by_value(augmented[i], 0, 255)
    plt.imshow(tf.cast(img, tf.uint8).numpy())
    plt.axis("off")
plt.suptitle("Augmented Samples (Flip + Rotation)")
plt.tight_layout()
plt.show()
```

        این متد یک batch از تصاویر دیتاست را دریافت کرده، لایه‌ی  augmentation را روی آن‌ها اعمال می‌کند و تعدادی از تصاویر را در قالب یک شبکه تصویری نمایش می‌دهد. این کار برای بررسی عملکرد لایه‌ی augmentation و اطمینان از درست بودن اعمال آن روی تصاویر انجام می‌شود. متد زیر را کامل کنید.

```python
# TODO: نمایش داده‌های Augmented
import math

# TODO: نمایش داده‌های Augmented
def visualize_augmented_images(dataset, augmentation_layer, num_samples=9):
    """
    نمایش تعدادی تصویر افزایش‌ یافته (Augmented) از یک دیتاست آموزشی.
    """
    # یک batch از دیتاست بگیر
    images, labels = next(iter(dataset))
    batch_size = images.shape[0]

    # تعداد نمونه‌هایی که نمایش می‌دهیم را به اندازه‌ی batch محدود کن
    n = min(num_samples, int(batch_size))

    # اعمال افزایش داده (training=True تا تبدیلات تصادفی فعال شوند)
    aug = augmentation_layer(images, training=True)

    # تبدیل ایمن برای نمایش (به uint8 در بازه‌ی 0..255)
    def to_uint8(x):
        x = tf.convert_to_tensor(x)
        # اگر داده‌ها در بازه‌ی 0..1 باشند، ضرب در 255
        xmax = tf.reduce_max(x)
        if tf.math.less_equal(xmax, 1.0):
            x = x * 255.0
        x = tf.clip_by_value(x, 0.0, 255.0)
        return tf.cast(x, tf.uint8)

    aug = to_uint8(aug)

    # ابعاد گرید (نزدیک مربع)
    cols = int(math.ceil(math.sqrt(n)))
    rows = int(math.ceil(n / cols))

    plt.figure(figsize=(3*cols, 3*rows))
    for i in range(n):
        ax = plt.subplot(rows, cols, i + 1)
        plt.imshow(aug[i].numpy())
        plt.axis('off')
    plt.suptitle("Augmented Samples", y=0.98)
    plt.tight_layout()
    plt.show()

visualize_augmented_images(train_ds, data_augmentation, num_samples=9)
```

        این بخش از باید پیش‌پردازش (Preprocessing) روی داده‌های تصویری قبل از ورود به مدل انجام دهید.
         در اینجا هدف اصلی نرمال‌سازی مقادیر پیکسل‌ها است، به این صورت که مقادیر اصلی که در بازه [0, 255] هستند، به بازه [0.0, 1.0] تبدیل می‌شوند.
         یک تابع تعریف شده که هر تصویر را از نوع int به float تبدیل می‌کند و سپس آن را بر ۲۵۵ تقسیم می‌کند تا نرمال شود. این تابع را کامل کنید و سپس  روی کل داده‌های آموزش و اعتبارسنجی با استفاده از متد map اعمال شده است.
    </li>
  </font>
<br><br>

```python
# تعریف تابع پیش‌پردازش شامل نرمال‌سازی مقادیر پیکسل
def preprocess(image, label):
    # تبدیل نوع داده به float32
    image = tf.cast(image, tf.float32)
    # نرمال‌سازی به بازه [0,1]
    image = image / 255.0
    return image, label

# اعمال روی دیتاست آموزش و اعتبارسنجی
train_ds = train_ds.map(preprocess, num_parallel_calls=tf.data.AUTOTUNE)
val_ds   = val_ds.map(preprocess, num_parallel_calls=tf.data.AUTOTUNE)
```

<br><br>
<div dir='rtl'>
  <h1><li>طراحی شبکه عصبی و اموزش مدل</li></h1>
  <font size=4>
    <li>
        در این مرحله،  باید یک مدل شبکه عصبی کانولوشنی (CNN) طراحی کنید که شامل لایه‌های استخراج ویژگی، کاهش ابعاد و لایه‌های تمام‌متصل برای طبقه‌بندی تصاویر به سه دسته مختلف باشد. همچنین، مدل باید کامپایل شده و آماده آموزش باشد.
    </li>
</font>
<br>

```python
# TODO: ساخت مدل شبکه عصبی کانولوشنی (CNN) برای طبقه‌بندی تصاویر

import tensorflow as tf
from tensorflow.keras import layers, models, optimizers

NUM_CLASSES = len(class_names)
INPUT_SHAPE = IMG_SIZE + (3,)

# --- Augmentation پیشنهادی قوی‌تر ---
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.2),
    layers.RandomZoom(0.15),
    layers.RandomContrast(0.15),
    layers.RandomTranslation(0.1, 0.1),
], name="data_augmentation")

# --- MobileNetV2 پایه (ImageNet) ---
base = tf.keras.applications.MobileNetV2(
    input_shape=INPUT_SHAPE, include_top=False, weights="imagenet"
)
base.trainable = False  # مرحله اول: فریز

# --- مدل نهایی ---
inputs = layers.Input(shape=INPUT_SHAPE)
x = data_augmentation(inputs)

# اگر تصاویرت از قبل در بازه [0,1] هستند، این لایه آن را به [-1,1] می‌برد
# (با این کار نیازی به mobilenet_v2.preprocess_input نیست و تداخلی هم با pipeline تو پیش نمی‌آید)
x = layers.Lambda(lambda t: (tf.cast(t, tf.float32) * 2.0) - 1.0, name="to_minus1_1")(x)

x = base(x, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.4)(x)
outputs = layers.Dense(NUM_CLASSES, activation="softmax")(x)

model = models.Model(inputs, outputs, name="rps_mobilenetv2_tl")

```

<br><br>
<div dir='rtl'>
    <font size=4>
        <li> مدل را کامپایل کنید و سپس آموزش دهید.
</div>
<br>

```python
# TODO: مدل را کامپایل کنید

model.compile(
    optimizer=optimizers.Adam(learning_rate=3e-4),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()
```

<br><br>
<div dir='rtl'>
  <font size=4>
    <li>
        مدل را آموزش دهید.
</font>

```python
# TODO: مدل را آموزش دهید

from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

callbacks = [
    EarlyStopping(monitor="val_loss", patience=3, restore_best_weights=True),
    ReduceLROnPlateau(monitor="val_loss", patience=2, factor=0.5, min_lr=1e-6),
]

EPOCHS = 12
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    callbacks=callbacks,
    verbose=1
)
```

```python
import matplotlib.pyplot as plt

def plot_training_history(history):
    """
    رسم گراف‌های دقت و خطا برای داده‌های آموزش و اعتبارسنجی.
    """
    acc      = history.history['accuracy']
    val_acc  = history.history['val_accuracy']
    loss     = history.history['loss']
    val_loss = history.history['val_loss']
    epochs   = range(1, len(acc) + 1)

    # --- نمودار دقت ---
    plt.figure(figsize=(12,5))

    plt.subplot(1,2,1)
    plt.plot(epochs, acc, 'b-', label='Training Accuracy')
    plt.plot(epochs, val_acc, 'r-', label='Validation Accuracy')
    plt.title('Accuracy over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)

    # --- نمودار خطا ---
    plt.subplot(1,2,2)
    plt.plot(epochs, loss, 'b-', label='Training Loss')
    plt.plot(epochs, val_loss, 'r-', label='Validation Loss')
    plt.title('Loss over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

history = model.fit(train_ds, validation_data=val_ds, epochs=30, callbacks=callbacks)
plot_training_history(history)
```
